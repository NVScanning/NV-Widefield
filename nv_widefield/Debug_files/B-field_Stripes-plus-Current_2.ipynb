```python
import numpy as np
import matplotlib.pyplot as plt
```

### Constants

```python
mu0 = 1.0
mu2pi = mu0/(2*np.pi)
```

## B-field generated by a single wire

```python
# Define the current strength and location
current = 10.0  # current strength mA
wire_pos_x = 0.0  # wire x-coordinate nm 
wire_pos_y = 0.0  # wire y-coordinate nm
```

```python
# Define the mesh grid of points, in nm
x, y = np.meshgrid(np.linspace(-200, 200, 200), np.linspace(-200, 200, 200))
```

```python
# Define the magnetic field components due to the current
r = np.sqrt((x-wire_pos_x)**2 + (y-wire_pos_y)**2)
BJ_x = mu2pi* current * (y-wire_pos_y) / r**3
BJ_y = -mu2pi * current * (x-wire_pos_x) / r**3
```

```python
# Plot the magnetic field lines
fig, ax = plt.subplots(figsize=(8, 8))
plt.scatter(wire_pos_x, wire_pos_y, c='red')
ax.streamplot(x, y, BJ_x, BJ_y, color='blue', linewidth=1, density=1, arrowstyle='->')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Plot the magnetic field components
center_element = x.shape[0]//2

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.title('B-field x (IP)')
plt.plot(x[0,:], BJ_x[:,center_element + 0], label='x = 0 nm') # cut through x
plt.plot(x[0,:], BJ_x[center_element + 0,:], label='y = 0 nm') # cut through y
plt.legend()

plt.subplot(122)
plt.title('B-field y (OOP)')
plt.plot(x[0,:], BJ_y[:,center_element + 0], label='x = 0 nm') # cut through x
plt.plot(x[0,:], BJ_y[center_element + 0,:], label='y = 0 nm') # cut through y

plt.legend()
plt.show()
```

```python
# Plot the magnetic field components
center_element = x.shape[0]//2

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.title('B-field x (IP)')
plt.ylim([-0.004, 0.004])
plt.plot(x[0,:], BJ_x[:,center_element + 10], label='x = 10 nm') # cut through x
plt.plot(x[0,:], BJ_x[center_element + 10,:], label='y = 10 nm') # cut through y
plt.plot(x[0,:], BJ_x[center_element + 20,:], label='y = 20 nm') # cut through y
plt.xlabel('x')
plt.legend()

plt.subplot(122)
plt.title('B-field y (OOP)')
plt.ylim([-0.004, 0.004])
plt.plot(x[0,:], BJ_y[:,center_element + 10], label='x = 10 nm') # cut through x
plt.plot(x[0,:], BJ_y[center_element + 10,:], label='y = 10 nm') # cut through y
plt.plot(x[0,:], BJ_y[center_element + 20,:], label='y = 20 nm') # cut through y
plt.xlabel('x')

plt.legend()
plt.show()
```

## B-field generated by an array of wires

```python
# wires coordinates
wire_pos_x = np.arange(-400, 401, 100)
# wire_pos_x = [0, 50]
wire_pos_y = 0.0 

print(wire_pos_x)
```

```python
# Define the mesh grid of points, in nm
xmin = -500
xmax = 500
nx = 500

ymin = -500
ymax = 500
ny = 500

x, y = np.meshgrid(np.linspace(xmin, xmax, nx), np.linspace(ymin, ymax, ny))
```

```python
# Define the magnetic field components due to the current
BJ_x = 0
BJ_y = 0
for pos_x in wire_pos_x:
    r = np.sqrt((x-pos_x)**2 + (y-wire_pos_y)**2)
    BJ_x += mu2pi* current * (y-wire_pos_y) / r**3
    BJ_y += -mu2pi * current * (x-pos_x) / r**3
    
```

```python
# Plot the magnetic field lines
fig, ax = plt.subplots(figsize=(8, 8))
plt.scatter(wire_pos_x, np.zeros(wire_pos_x.shape), c='red')
ax.streamplot(x, y, BJ_x, BJ_y, color='blue', linewidth=1, density=1, arrowstyle='->')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Plot the magnetic field components
center_element = x.shape[0]//2

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.title('B-field x (IP)')
plt.plot(x[0,:], BJ_x[:,center_element + 0], label='x = 0 nm') # cut through x
plt.plot(x[0,:], BJ_x[center_element + 0,:], label='y = 0 nm') # cut through y
plt.legend()

plt.subplot(122)
plt.title('B-field y (OOP)')
plt.plot(x[0,:], BJ_y[:,center_element + 0], label='x = 0 nm') # cut through x
plt.plot(x[0,:], BJ_y[center_element + 0,:], label='y = 0 nm') # cut through y

plt.legend()
plt.show()
```

```python
# Plot the magnetic field components
center_element = x.shape[0]//2

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.title('B-field x (IP)')
plt.ylim([-0.004, 0.004])
plt.plot(x[0,:], BJ_x[:,center_element + 10], label='x = 10 nm') # cut through x
plt.plot(x[0,:], BJ_x[center_element + 10,:], label='y = 10 nm') # cut through y
plt.plot(x[0,:], BJ_x[center_element + 20,:], label='y = 20 nm') # cut through y
plt.xlabel('x')
plt.legend()

plt.subplot(122)
plt.title('B-field y (OOP)')
plt.ylim([-0.004, 0.004])
plt.plot(x[0,:], BJ_y[:,center_element + 10], label='x = 10 nm') # cut through x
plt.plot(x[0,:], BJ_y[center_element + 10,:], label='y = 10 nm') # cut through y
plt.plot(x[0,:], BJ_y[center_element + 20,:], label='y = 20 nm') # cut through y
plt.xlabel('x')

plt.legend()
plt.show()
```

## B-field generated by a single magnet

```python
# Define the magnetic moment strength and location
moment = [0, 1.0]  # current strength mA
moment_pos_x = 0.0  # wire x-coordinate nm 
moment_pos_y = 0.0  # wire y-coordinate nm
```

```python
# Define the mesh grid of points, in nm
x, y = np.meshgrid(np.linspace(-200, 200, 200), np.linspace(-200, 200, 200))
```

```python
# Define the magnetic field components due to the magnetic moment
r = np.sqrt((x-moment_pos_x)**2 + (y-moment_pos_y)**2)
r3 = r**3
r5 = r3 * r * r
mdotr = moment[0] * (x - moment_pos_x) + moment[1] * (y - moment_pos_y)

BM_x = mu2pi * 0.5 * (3 * mdotr * (x - moment_pos_x) / r5 - moment[0] / r3)
BM_y = mu2pi * 0.5 * (3 * mdotr * (y - moment_pos_y) / r5 - moment[1] / r3)
```

```python
# Plot the magnetic field lines
fig, ax = plt.subplots(figsize=(8, 8))
plt.scatter(moment_pos_x, moment_pos_y, c='green')
ax.streamplot(x, y, BM_x, BM_y, color='blue', linewidth=1, density=1, arrowstyle='->')
plt.xlabel('x')
plt.ylabel('y')
```
